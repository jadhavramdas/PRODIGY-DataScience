```python
df=df.drop(['Cabin', 'Name', 'PassengerId', 'Ticket'], axis=1)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegressionCV
from sklearn.metrics import confusion_matrix,classification_report
```

```python
X=pd.get_dummies(df.drop('Survived', axis=1), drop_first=True)
y=df['Survived']
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
model = RandomForestClassifier(n_estimators=10,max_features=10,random_state=101)
```

```python
model.fit(X_train,y_train)
```

```python
preds = model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score,ConfusionMatrixDisplay
```

```python
cm=confusion_matrix(y_test,preds)
```

```python
disp = ConfusionMatrixDisplay(confusion_matrix=cm,
                              display_labels=model.classes_)
disp.plot()
```

```python
print(classification_report(y_test,preds))
```

```python
features=pd.DataFrame(index=X.columns,data=model.feature_importances_,columns=['Feature Importance'])
features.sort_values(by='Feature Importance')
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
n_estimators=[20,40,60,80]
max_features= [4,5,6]
bootstrap = [True,False]
oob_score = [True,False]
```

```python
param_grid = {'n_estimators':n_estimators,
             'max_features':max_features,
             'bootstrap':bootstrap,
             'oob_score':oob_score}
```

```python
rfc = RandomForestClassifier()
grid = GridSearchCV(rfc,param_grid)
```

```python
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
predictions = grid.predict(X_test)
```

```python
print(classification_report(y_test,predictions))
```
